```python
main_df = pd.read_csv("/content/2023_Q3.2_AllData_Before_20231215 2.csv", header=0, index_col=0, low_memory=False).reset_index(drop=True)
```

```python
new_df = pd.DataFrame(main_df[["SeriesCode","SeriesDescription"]].drop_duplicates())
new_df.set_index("SeriesCode", inplace=True)
```

```python
print(new_df)
```

```python
# summarized series description

new_df["SeriesDescription"].describe()
```

```python
# Applying NLPTK to Series Description

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')

def process_text(text):
  # Tokenize the text
  tokens = nltk.word_tokenize(text)

  # Remove stop words
  stop_words = set(nltk.corpus.stopwords.words('english'))
  tokens = [token for token in tokens if token.lower() not in stop_words]

  # Lemmatize the remaining words
  lemmatizer = nltk.stem.WordNetLemmatizer()
  tokens = [lemmatizer.lemmatize(token) for token in tokens]

  return tokens

new_df["SeriesDescription_processed"] = new_df["SeriesDescription"].apply(process_text)

new_df.head()
```

```python
# remove punctuation in series description

import string

def remove_punctuation(text):
  # Create a translation table to map punctuation characters to None
  translator = str.maketrans('', '', string.punctuation)

  # Use the translation table to remove punctuation from the text
  return text.translate(translator)

new_df["SeriesDescription_processed"] = new_df["SeriesDescription"].apply(remove_punctuation)

# Show the first five rows of the new DataFrame
new_df.head()
```

```python
# Using dataframe new_df: Summarizing text in series description

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(new_df['SeriesDescription'])
tfidf_matrix = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_matrix.mean().sort_values(ascending=False)
```

```python
# shortening and summarizing the text in series descriptions

# Import the necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download the necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Define a function to shorten and summarize text
def summarize_text(text, max_length=100):
  # Tokenize the text
  tokens = nltk.word_tokenize(text)

  # Remove stop words
  stop_words = set(stopwords.words('english'))
  tokens = [token for token in tokens if token.lower() not in stop_words]

  # Stem the tokens
  stemmer = PorterStemmer()
  tokens = [stemmer.stem(token) for token in tokens]

  # Join the tokens back into a string
  summary = ' '.join(tokens)

  # Truncate the summary to the desired length
  if len(summary) > max_length:
    summary = summary[:max_length] + ''

  return summary

# Apply the summarize_text function to the SeriesDescription column
new_df['SeriesDescription_summarized'] = new_df['SeriesDescription_processed'].apply(summarize_text)

# Print the first five rows of the new DataFrame
new_df.head()
```

```python
# removing ilo from series description summarrized

new_df['SeriesDescription_summarized'] = new_df['SeriesDescription_summarized'].apply(lambda x: x.replace('ilo', ''))
```
