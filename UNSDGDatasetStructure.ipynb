```python
import pandas as pd
import numpy as np
from termcolor import colored

# Load your data
data = pd.read_csv('2023_Q3.2_AllData_Before_20231215.csv', header=0, index_col=0, low_memory=False).reset_index(drop=True)
```

```python
pd.DataFrame.iteritems = pd.DataFrame.items

for column_name, column_values in data.iteritems():
    print(colored(f"Field: \"{column_name}\"", attrs=['bold']))
    print(colored(f"Count of Unique Values: {data[column_name].nunique()}", attrs=['bold']))
    print(data[column_name].unique())
    print(colored(f"Number of records for each \"{column_name}\" : ", attrs=['bold']))
    print(data[column_name].value_counts())
```

```python
%%capture cap

df2 = data   #.replace(np.nan,'NaN', regex=True)
pd.DataFrame.iteritems = pd.DataFrame.items
for column_name, column_values in df2.iteritems():
    print(f"Field: \" {column_name} \" ")
    print(f"Count of Unique Values: {df2[column_name].nunique()}")
    print(df2[column_name].unique())
    print(f"Number of records for each \"{column_name}\" : ")
    print(df2[column_name].value_counts().sort_values, '\n')

with open('SDGDatasetVariables.txt', 'w', encoding='utf-8') as file:
    file.write(cap.stdout)
```
